```
from pyspark.sql.functions import explode,trim,regexp_replace,lower,col,desc,split;
```

```
#Function for groupby the dataframe
def wordcount(col):
        return (col.groupby(col.Col_explode).count())
```

```
#Function for regexp inorder to replace the symbools in a given string
def remove_puncations(values):
    return trim(lower(regexp_replace(values,'([^\s\w_]|_)+',''))).alias('sentence')
```

```
# Read the text file using spark#
df=spark.read.text("dbfs:/FileStore/The_Project_Gutenberg_eBook_of_The_.txt").select(remove_puncations(col('value')))
df.show(15,truncate=False)
```

```
#Split the resultant values of regexp dataframe
split=(df.select(split(df.sentence, '\s+').alias('Col_split')))

#Explode the resultant values of split dataframe
explode=(split.select(explode(split.Col_split).alias('Col_explode')))

#Retsricting the notnull values
notnull=explode.where(explode.Col_explode.isNotNull())

#Count of Words in a resultant dataframe
word_count=wordcount(notnull)

# Dispaly the reuslts
display(word_count)
```

```
#orderBy the Dataframe count asc
order=word_count.orderBy("count", ascending=True)
display(order)
```
